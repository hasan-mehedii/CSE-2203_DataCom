```python
import matplotlib.pyplot as plt
```

```python
def ami_i_encode(input_str):
  encoded = []
  last = -1
  for bit in input_str:
        if bit == '0':
            encoded.append(0)
        elif bit == '1':
            last *= -1
            encoded.append(last)
        else:
            print("Invalid bit")
  return encoded
```

```python
def ami_i_decode(voltage_list):
  binary_str = ''
  for str in voltage_list:
    if str == 0:
      binary_str += '0'
    elif str == 1 or str == -1:
      binary_str += '1'
    else:
      print("Value error")
  return binary_str 
```

```python
def plot_signal(encoded_list):
  time = list(range(len(encoded_list) + 1))
  signal = [encoded_list[0]] + encoded_list
  plt.step(time, signal, where='post')
  plt.ylim(-2, 2)
  plt.xlabel('Time')
  plt.ylabel('Voltage Level')
  plt.title('AMI-I Encoding Signal')
  plt.grid(True)
  plt.show()
```

```python
input_str = input()
```

```python
en = ami_i_encode(input_str)
print(en)
plot_signal(en)
```

```python
dc = ami_i_decode(en)
print(dc)
```
